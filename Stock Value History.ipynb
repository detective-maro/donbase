```python
import requests

def get_stock_symbol(company_name):
    # Alpha Vantage API endpoint for symbol search
    endpoint = "https://www.alphavantage.co/query"
    
    # API key (replace 'YOUR_API_KEY' with your actual API key)
    api_key = '11XL0HUND8OAPVYS'
    
    # Parameters for the API request
    params = {
        "function": "SYMBOL_SEARCH",
        "keywords": 'company_name',
        "apikey": api_key
    }
    
    # Make the API request
    response = requests.get(endpoint, params=params)
    data = response.json()
    
    # Check if the response contains search results
    if 'bestMatches' in data:
        # Extract the first search result (assuming it's the most relevant)
        first_result = data['bestMatches'][0]
        symbol = first_result['1. symbol']
        return symbol
    else:
        return None

def get_stock_high_low(symbol):
    # Alpha Vantage API endpoint for stock time series
    endpoint = "https://www.alphavantage.co/query"
    
    # API key (replace 'YOUR_API_KEY' with your actual API key)
    api_key = 'YOUR_API_KEY'
    
    # Parameters for the API request
    params = {
        "function": "TIME_SERIES_DAILY",
        "symbol": symbol,
        "apikey": api_key
    }
    
    # Make the API request
    response = requests.get(endpoint, params=params)
    data = response.json()
    
    # Check if the response contains stock data
    if 'Time Series (Daily)' in data:
        # Extract the latest stock data
        latest_data = data['Time Series (Daily)']
        latest_date = list(latest_data.keys())[0]
        high = float(latest_data[latest_date]['2. high'])
        low = float(latest_data[latest_date]['3. low'])
        return high, low
    else:
        return None, None

# Example usage
company_name = 'Rivian Automotive Inc.'  # Example company name
symbol = get_stock_symbol(company_name)
if symbol:
    high, low = get_stock_high_low(symbol)
    if high is not None and low is not None:
        print(f"Stock High for {company_name} ({symbol}): {high}")
        print(f"Stock Low for {company_name} ({symbol}): {low}")
    else:
        print(f"Failed to fetch stock data for {company_name} ({symbol}).")
else:
    print(f"No stock symbol found for {company_name}. Please check the company name or try again later.")
```

```python
import requests
from datetime import datetime, timedelta

def get_stock_high_low(symbol, start_date, end_date):
    # Alpha Vantage API endpoint for stock time series
    endpoint = "https://www.alphavantage.co/query"
    
    # API key (replace 'YOUR_API_KEY' with your actual API key)
    api_key = '11XL0HUND8OAPVYS'
    
    # Parameters for the API request
    params = {
        "function": "TIME_SERIES_DAILY",
        "symbol": symbol,
        "apikey": api_key
    }
    
    # Calculate the number of days between start_date and end_date
    delta = end_date - start_date
    days = delta.days
    
    # Adjust the number of data points based on the desired time range
    if days < 100:
        params["outputsize"] = "compact"  # Return only the last 100 data points
    else:
        params["outputsize"] = "full"  # Return full historical data
    
    # Make the API request
    response = requests.get(endpoint, params=params)
    data = response.json()
    
    # Check if the response contains stock data
    if 'Time Series (Daily)' in data:
        # Extract stock data within the specified date range
        stock_data = {}
        for date, values in data['Time Series (Daily)'].items():
            date_obj = datetime.strptime(date, "%Y-%m-%d")
            if start_date <= date_obj <= end_date:
                stock_data[date_obj] = {
                    'high': float(values['2. high']),
                    'low': float(values['3. low'])
                }
        return stock_data
    else:
        print("Error: No stock data found.")
        return None

# Example usage
symbol = 'AAPL'  # Example stock symbol
end_date = datetime.now()
start_date = end_date - timedelta(days=365)  # One year ago
stock_data = get_stock_high_low(symbol, start_date, end_date)
if stock_data:
    print(f"Stock Data for {symbol} from {start_date.strftime('%Y-%m-%d')} to {end_date.strftime('%Y-%m-%d')}:\n")
    for date, values in stock_data.items():
        print(f"Date: {date.strftime('%Y-%m-%d')} - High: {values['high']} - Low: {values['low']}")
else:
    print("Failed to fetch stock data. Please check the symbol or try again later.")
```
